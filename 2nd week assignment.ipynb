## Questions
<ol><li>Read the birth_life_2010.txt file line by line</li>
<li>Make sure that every country name is in the same list as the Birth rate and the Life expectancy belonging to that country</li>
<li>Make sure all the countries names starts from capital letters</li>
<li>Round all the numbers to 2 decimals</li>
<li>Sort the list based on the country names from A -> Z</li>
<li>Remove now the last letter of every country</li>

```python
with open("data/birth_life_2010.txt","r",encoding='utf8') as f:
    text=f.read().split('\n')
    countries=[item.split('\t') for item in text]
    
```

```python
countries[:3]
```

```python
capitalized_countries=[]
for item in countries[1:]:
    capitalized_countries.append([item[0].capitalize(),round(float(item[1]), 2), round(float(item[2]), 2)])
sorted(capitalized_countries[1:])
```

```python
new_countries = []
for item in capitalized_countries[1:]:
    new_countries.append([item[0][:-1],item[1],item[2]])
    
sorted(new_countries)
```

## Hard Questions
<ol><li>Create a function that takes in a list of words and returns step 3 up to step 6 as a list</li>
<li>How many countries does the file contain?</li>
<li>Do all countries have birth and life expectancy present?</li>
<li>How many countries start with the letter M?</li>
<li>How many countries have a 'z' in their name</li>

```python
def transform_this(countries):
    capitalized_countries = []
    for item in countries[1:]:
        capitalized_countries.append([item[0][:-1].capitalize(), round(float(item[1]), 2), round(float(item[2]), 2)])
    return sorted(capitalized_countries)

transform_this(countries)
```
